1. State the value of $b$, given the equation of a curve is $y = -x^{2} + bx + c$

  
The roots from the curve are $x = [-2, 22]$  
$y = -(x + 2)(x - 22)$  
$y = -(x^{2} + 2x - 22x - 44)$  
$y = -x^{2} - 2x + 22x + 44$  
$y = -x^{2} + 20x + 44$  
$b = 20$

2. For Q1, state the value of $c$.

$c = 44$

3. The quadratic equation $x^{2} + kx + k = 1$ has equal roots. Find $k$.

$a = 1$, $b = k$, $c = k - 1$  
$\Delta _{2} = b^{2} - 4ac = 0$  
$= k^{2} - 4 \times 1 \times (k - 1) = 0$  
$= k^{2} - 4k - 4 = 0$  
$= (k - 2)^{2} - 8 = 0$  
$k = 2$
